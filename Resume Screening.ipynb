Cleaning Data : 
1. URLs,
2. Hashtags,
3. mentions,
4. special letters,
5. punctuations.

```python
import re
def cleanResume(txt):
    cleanTxt = re.sub('http\S+\s',' ',txt)   #Removes words starting with http
    cleanTxt = re.sub('@\S+',' ',cleanTxt) 
    cleanTxt = re.sub('#\S+',' ',cleanTxt)
    cleanTxt = re.sub('RT|cc',' ',cleanTxt)
    cleanTxt = re.sub('[%s]' % re.escape("""!"#$%&'()*+,./:;<=>?@[\]^-_`{|}~"""),' ',cleanTxt)
    cleanTxt = re.sub(r'[^\x00-\x7f]',' ',cleanTxt)
    cleanTxt = re.sub('\s+',' ',cleanTxt)
    return cleanTxt
```

```python
cleanResume("my website link is this http://helloword and % @ $ access it hi@gmail.com")
```

```python
txt = df['Resume'][0]
```

```python
cleanResume(txt)
```

```python
df['Resume'] = df['Resume'].apply(lambda x: cleanResume(x))
```

```python
df['Resume'][0]
```

Words into categorical values

```python

from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

```python
le.fit(df['Category'])
df['Category'] = le.transform(df['Category'])
```

```python
print(df['Category'])
```

Vectorization

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tfidf = TfidfVectorizer(stop_words='english')

tfidf.fit(df['Resume'])
requiredText = tfidf.transform(df['Resume'])
```

```python
requiredText
```

Splitting


```python
from sklearn.model_selection import train_test_split
```

```python
x_train , x_test, y_train ,y_test = train_test_split(requiredText,df['Category'],test_size=0.33,random_state=42)
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.multiclass import OneVsRestClassifier
from sklearn.metrics import accuracy_score

clf = OneVsRestClassifier(KNeighborsClassifier())
clf.fit(x_train,y_train)
y_pred = clf.predict(x_test)
print(accuracy_score(y_test,y_pred))
```
